Для $U[0, 1]$ сгенерируйте $200$ независимых случайных величин, постройте график плотности на отрезке $[-0.25, 1.25]$, а также гистограмму по сгенерированным случайным величинам. Для построения графика необходимо брать достаточно большое количество точек сетки.

```python
#имплементация из контеста
coin = sps.bernoulli(p=0.5).rvs


def uniform(size=1, precision=30):
    return (np.tensordot(coin((np.prod(size), precision)), 2. ** np.arange(-precision, 0), axes=([1], [0]))).reshape(size)

def normal(size=1, loc=0, scale=1, precision=30):
      return loc + scale * np.cos(2.0 * np.pi * uniform(size, precision)) * np.sqrt(-2.0 * np.log(uniform(size, precision)))

def expon(size=1, lambd=1, precision=30):
    return -1 * np.log(uniform(size, precision)) / lambd
```

```python
#генерация рандомных данных
u_data = uniform(size=200)

plt.figure(figsize=(10, 5))
plt.hist(u_data, bins=100, density=True, alpha=0.6, label='Сгенерированные данные')
x = np.linspace(-0.25, 1.25, 1000)
plt.plot(x, sps.uniform.pdf(x), 'red', lw=2, label='Теоретическая плотность')
plt.title('Равномерное распределение U[0,1]')
plt.xlabel('Случайная величина')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True)
plt.show()
```

Исследуйте, как меняются значения случайных величин в зависимости от `precision` в написанной ранее вами функции.

```python
plt.figure(figsize=(15, 3))
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
precisions = [0, 1, 2, 5, 10, 20]

for i, p in enumerate(precisions):
    ax = plt.subplot(1, 6, i+1)
    samples = uniform(size=100, precision=p)
    ax.scatter(samples, np.zeros(100), alpha=0.4, color=colors[i], label=f'точность={p}')
    ax.set_yticks([])
    ax.set_title(f'Точность {p}%')
    ax.set_xlim(-0.1, 1.1)

plt.suptitle('Зависимость распределения от точности', y=1.1)
plt.tight_layout()
plt.show()
```

Для $\mathcal{N}(0, 1)$ сгенерируйте $200$ независимых случайных величин, постройте график плотности на отрезке $[-3, 3]$, а также гистограмму по сгенерированным случайным величинам.

```python
normal_data = normal(size=200)

plt.figure(figsize=(10, 5))
plt.hist(normal_data, bins=100, density=True, alpha=0.6, label='Сгенерированные данные')
x = np.linspace(-3, 3, 1000)
plt.plot(x, sps.norm.pdf(x), 'red', lw=2, label='Теоретическая плотность')
plt.title('Нормальное распределение N(0,1)')
plt.xlabel('Случайная величина')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True)
plt.show()
```

Для $Exp(1)$ сгенерируйте выборку размера 100 и постройте график плотности этого распределения на отрезке $[-0.5, 5]$.

```python
exp_data = expon(100)

plt.figure(figsize=(10, 5))
plt.hist(exp_data, bins=100, density=True, alpha=0.6, label='Сгенерированные данные')
x = np.linspace(-0.5, 5, 1000)
plt.plot(x, sps.expon.pdf(x), 'red', lw=2, label='Теоретическая плотность')
plt.title('Экспоненциальное распределение Exp(1)')
plt.xlabel('Случайная величина')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True)
plt.show()
```

Рассмотрим <a target="_black" href="https://ru.wikipedia.org/wiki/Распределение_Коши">распределение Коши</a> с плотностью
$$p(x) = \frac{1}{\pi(1+x^2)}$$
для $x\in\mathbb{R}$, которое известно своими <a target="_black" href="https://en.wikipedia.org/wiki/Heavy-tailed_distribution">тяжелыми хвостами</a>. Иначе говоря, плотность убывает медленно при $x \to \infty$ (точное определение может различаться для разных источников). Сравните, например, с нормальным или экспоненциальным распределениями. В частности, из-за этого у распределения Коши даже не существует математического ожидания.

За распределение Коши отвечает класс <a target="_black" href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.cauchy.html">`scipy.stats.cauchy`</a>.

**1.** Визуализируйте на одном графики плотности нормального распределения и распределения Коши. Что можно сказать?

```python
grid = np.linspace(-3, 3, 1000)
cauchy_data = sps.cauchy.pdf(grid)

plt.figure(figsize=(10,5))
plt.plot(grid, cauchy_data, linewidth=5, color='green', label='Распределение Коши')
plt.plot(grid, sps.norm.pdf(grid), linewidth=5, color='red', label='Нормальное распределение')

plt.legend()
plt.xlabel('Случайная величина')
plt.ylabel('Плотность')
plt.title('Зависимость плотности от значения случайной величины')
```

В нуле нормальное распределние больше нормальное распределение, однако оно и убывает гораздо быстрее

**2.** Повторите эксперимент с <a href="https://thetahat.ru/courses/ad/main/9/lec9_LLN" target="_top">занятия</a> про закон больших чисел для распределения Коши. Выполняется ли закон?

```python
size = 1000  # количество случайных величин
samples_count = 500  # количество выборок

samples = sps.cauchy.rvs(size=(samples_count, size))
cum_means = samples.cumsum(axis=1) / (np.arange(size) + 1)
```

```python
plt.figure(figsize=(15, 7))

# рисуем для каждой выборки отдельно
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, cum_means[i], color="green", alpha=0.01)

plt.xlabel("Количество случайных величин")
plt.ylabel("Значение среднего")
plt.xlim((0, size));
y_limit = 20
plt.ylim(-y_limit, y_limit)
plt.show()
```

Как мы видим для распределения Коши не выполнено ЗБЧ(График несколько блуждает). Это было предсказуемо, ведь у распределения Коши нет мат.ожидания.

**3.** Аналогичным образом проведите эксперимент по изучению зависимости *выборочной* медианы от размера выборки. Медиану по выборке можно посчитать с помощью <a target="_blank" href="https://numpy.org/doc/stable/reference/generated/numpy.median.html">`np.median`</a>.

```python
def cumulative_median(array):
    return [np.median(array[:k+1]) for k in range(len(array))]

cum_median = np.apply_along_axis(cumulative_median, axis=1, arr=samples)
```

```python
plt.figure(figsize=(15, 7))

# рисуем для каждой выборки отдельно
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, cum_median[i], color="green", alpha=0.01)

plt.xlabel("Количество случайных величин")
plt.ylabel("Значение среднего")
plt.xlim((0, size));
y_limit = 5
plt.ylim(-y_limit, y_limit)
plt.show()

```

Как видим, ЗБЧ тут работает

**Вывод:**

В ходе работы выяснилось: \\
1) для распределения Коши ЗБЧ не работает
2) для медиан распрпделения Коши ЗБЧ работает видимо потому, что медиана более устойчива к выбросам т.к сильнее зависит от центральной части распределения, а значит у нее примерно нормальное распределение
